```python
import numpy as np
import matplotlib.pyplot as plt

from itertools import cycle
x = np.arange(20)

colors = ['yellow','black','red']
color_gen = cycle(colors)

for i in range(1, 9):
    plt.scatter(x, i * x**3, label='Group %d' % i, color=next(color_gen))

plt.legend(loc='upper left')
plt.show()
```

```python
from itertools import cycle
x = np.arange(20)

colors = ['blue', 'red', 'green']
color_gen = cycle(colors)

for i in range(1, 9):
    plt.plot(x, i * x**3, label='Group %d' % i, marker='*')

plt.legend(loc='upper left')
plt.show()
```
